```python
from matplotlib import pyplot as plt
```

# Line Charts

```python
years= [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9 , 2862.5, 5989.6, 10289.7, 14958.3]

# create a line char, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color = 'green', marker = 'o' , linestyle = 'solid')

#add a title
plt.title("Nominal GDP")

#add a lable to the y-axis
plt.ylabel("Billions of dollars")
plt.show()
```

# Bar charts

```python
movies = ["Raging Bull", "Ben Hur", "Casablanca", "Fellowship of the ring", "West Side Story"]
num_oscars = [2, 11, 3, 4 , 10]

#plot bars with left x-cordinates [0,1,2,3,4] , heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("My favorie movies")
plt.ylabel("# of academy awards")

#label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

#Histogram

```python
 from collections import Counter
grades = [83, 95, 91, 87, 70, 0 , 85, 82, 100, 67, 73, 77, 0]

#Bucket grades by decile , but put 100 in the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()] , 
        histogram.values(),
        10,
        edgecolor = (0, 0, 0))

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("no of students")
plt.title("Distribution of exam 1 grades")
plt.show()
```

# Trends

```python
variance = [1,2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128 ,64 , 32, 16, 8 , 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

#make multiple calls to pyplot.plot
# to show multiple series on the same chart
plt.plot(xs, variance, 'g-', label= 'variance')
plt.plot(xs, bias_squared, 'r-', label= 'bias^2')
plt.plot(xs, total_error, 'b:', label= 'total error')

# Because we've assigned lables to each series,
#we can get a legend for free at top center
plt.legend(loc= 9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The bias variance tradeoff")
plt.show()
```

# scatter plots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

#label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
  plt.annotate(label,
               xy= (friend_count, minute_count),
               xytext = (5, -5),
               textcoords = 'offset points')
  
plt.title("Daily Minutes vs Num of friends")
plt.xlabel('# of friends')
plt.ylabel('daily minutes spent on on the site')
plt.show()
```

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("axes aren't comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

```python
plt.scatter(test_1_grades, test_2_grades)
plt.axis("equal") # to show an equal axis
plt.title("axes are comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```
