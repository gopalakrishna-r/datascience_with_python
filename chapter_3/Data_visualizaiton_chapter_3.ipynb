```
from matplotlib import pyplot as plt
```

Line chart

```
years= [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9 , 2862.5, 5989.6, 10289.7, 14958.3]

# create a line char, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color = 'green', marker = 'o' , linestyle = 'solid')

#add a title
plt.title("Nominal GDP")

#add a lable to the y-axis
plt.ylabel("Billions of dollars")
plt.show()
```

Bar chart

```
movies = ["Raging Bull", "Ben Hur", "Casablanca", "Fellowship of the ring", "West Side Story"]
num_oscars = [2, 11, 3, 4 , 10]

#plot bars with left x-cordinates [0,1,2,3,4] , heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("My favorie movies")
plt.ylabel("# of academy awards")

#label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

histogram

```
from collections import Counter
grades = [83, 95, 91, 87, 70, 0 , 85, 82, 100, 67, 73, 77, 0]

#Bucket grades by decile , but put 100 in the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

print(histogram)

plt.bar([x + 5 for x in histogram.keys()] , 
        histogram.values(),
        10,
        edgecolor = (0, 0, 0))

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("no of students")
plt.title("Distribution of exam 1 grades")
plt.show()
```
